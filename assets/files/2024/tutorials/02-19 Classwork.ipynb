```python
import table_rl
import numpy as np

env = table_rl.envs.BasicEnv(discount=0.9)
num_states = 3
num_actions = 2 # L and R
T = env.T
R = env.R
discount = 0.9

explorer = table_rl.explorers.ConstantEpsilonGreedy(0.1, T.shape[1])

agent = table_rl.learners.QLearning(T.shape[0],
                  T.shape[1],
                  0.02,
                  explorer,
                  discount=discount,
                  initial_val=0.)

observation, info = env.reset()

for _ in range(100):
    action = agent.act(observation, True)
    observation, reward, terminated, truncated, info = env.step(action)
    agent.observe(observation, reward, terminated, truncated)
    if terminated or truncated:
        observation, info = env.reset()

hand_confirmed_opt = np.array([[0.79345359, 0.8708637 ],
                                [0.80539959, 0.97837773],
                                [0.,         0.        ]])

print(agent.q)
```
