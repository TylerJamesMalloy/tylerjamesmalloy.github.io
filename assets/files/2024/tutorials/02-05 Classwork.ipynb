```python
from pyibl import Agent
a = Agent(name = "Ty", default_utility=0.5)

choice = a.choose(["Left", "Middle", "Right"], details=True)
print(choice)
# ('Left',
# [{'choice': 'Left', 'blended_value': 0.5, 'retrieval_probabilities': []},
#  {'choice': 'Right', 'blended_value': 0.5, 'retrieval_probabilities': []}])

if(choice[0] == "Left"):
  a.respond(0.2)
if(choice[0] == "Right"):
  a.respond(0.8)
if(choice[0] == "Middle"):
  a.respond(0.8)


choice = a.choose(["Left", "Right"], details=True)
print(choice)
# ('Right', [
# {'choice': 'Right', 'blended_value': 0.5, 'retrieval_probabilities': []},
# {'choice': 'Left', 'blended_value': 0.2, 'retrieval_probabilities':
  #[{'utility': 0.2, 'retrieval_probability': 1.0}]}])


if(choice[0] == "Left"):
  a.respond(0.4)
if(choice[0] == "Right"):
  a.respond(0.6)

choice = a.choose(["Left", "Middle"], details=True)
print(choice)

# ('Right', [
# {'choice': 'Right', 'blended_value': 0.6, 'retrieval_probabilities':
 # [{'utility': 0.6, 'retrieval_probability': 1.0}]},
# {'choice': 'Left', 'blended_value': 0.2, 'retrieval_probabilities':
 # [{'utility': 0.2, 'retrieval_probability': 1.0}]}])

if(choice[0] == "Left"):
  a.respond(0.6)
if(choice[0] == "Middle"):
  a.respond(0.4)

choice = a.choose(["Left", "Right"], details=True)
print(choice)

# ('Right', [
# {'choice': 'Right', 'blended_value': 0.5677693878339869, 'retrieval_probabilities':
 #  [{'utility': 0.6, 'retrieval_probability': 0.8388469391699344},
 #   {'utility': 0.4, 'retrieval_probability': 0.16115306083006561}]},
# {'choice': 'Left', 'blended_value': 0.2, 'retrieval_probabilities':
 # [{'utility': 0.2, 'retrieval_probability': 1.0}]}])


if(choice[0] == "Left"):
  a.respond(0.8)
if(choice[0] == "Right"):
  a.respond(0.2)


choice = a.choose(["Left", "Right"], details=True)
print(choice)

# ('Right', [
# {'choice': 'Right', 'blended_value': 0.5281514076246109, 'retrieval_probabilities':
 # [{'utility': 0.6, 'retrieval_probability': 0.8002534318478912},
  # {'utility': 0.4, 'retrieval_probability': 0.04025017442727174},
   # {'utility': 0.2, 'retrieval_probability': 0.1594963937248371}]},
# {'choice': 'Left', 'blended_value': 0.2, 'retrieval_probabilities':
 # [{'utility': 0.2, 'retrieval_probability': 1.0}]}])

if(choice[0] == "Left"):
  a.respond(0.9)
if(choice[0] == "Right"):
  a.respond(0.1)

choice = a.choose(["Left", "Right"], details=True)
print(choice)

# ('Right', [
# {'choice': 'Right', 'blended_value': 0.20142491116657604, 'retrieval_probabilities':
 # [{'utility': 0.6, 'retrieval_probability': 0.00947848064381874},
  # {'utility': 0.4, 'retrieval_probability': 0.031202319375770244},
   # {'utility': 0.2, 'retrieval_probability': 0.08732497503193559},
    # {'utility': 0.1, 'retrieval_probability': 0.8606944966105515}]},
# {'choice': 'Left', 'blended_value': 0.2, 'retrieval_probabilities':
 # [{'utility': 0.2, 'retrieval_probability': 1.0}]}])

if(choice[0] == "Left"):
  a.respond(1.0)
if(choice[0] == "Right"):
  a.respond(0)

choice = a.choose(["Left", "Right"], details=True)
print(choice)

# ('Left', [
# {'choice': 'Left', 'blended_value': 0.2, 'retrieval_probabilities':
 # [{'utility': 0.2, 'retrieval_probability': 1.0}]},
# {'choice': 'Right', 'blended_value': 0.12511229001634352, 'retrieval_probabilities':
 # [{'utility': 0.6, 'retrieval_probability': 0.030277797433255586},
  # {'utility': 0.4, 'retrieval_probability': 0.058568254158452356},
   # {'utility': 0.2, 'retrieval_probability': 0.06537893906858536},
    # {'utility': 0.1, 'retrieval_probability': 0.7044252207929215},
     # {'utility': 0, 'retrieval_probability': 0.14134978854678526}]}])
```
